```python
import numpy as np
from IPython.display import display, HTML
```

```python
data = [
    "BLDP,0.10,2.00",
    "CJT,0.11,4.00",
    "AAPL,0.80,3.00",
]
labels = []
weights = []
values = []
items = 0
capacity = 1*100
```

```python
for line in data:
    symbol, price, target = line.split(",")
    labels.append(symbol)
    weights.append(int(float(price)*100))
    values.append(int(float(target)*100))
    items +=1
```

```python
grid = np.empty((items + 1, capacity + 1))
grid[0] = 0

for item in range(items):
    print(grid[item])
    print(grid[item+1])
    this_weight = weights[item]
    this_value = values[item]
    grid[item+1, :this_weight] = grid[item, :this_weight] # for k amount of items, amassing up to the weight of the current item, duplicate data from k-1 items
    print(grid[item])
    print(grid[item+1])
    temp = grid[item, :-this_weight] + this_value # for k amount of items, amassing up to (max capacity - the weight of the current item) equate to data from k-1 items, adding on weight of current item
    print(temp)

    grid[item + 1, this_weight:] = np.where(temp > grid[item, this_weight:], temp, grid[item, this_weight:])
    print(grid[item])
    print(grid[item+1])
    print("\n\n")
```

```python
for item_amount, weightss in enumerate(grid):
    print("\n" + str(item_amount) + "-" + data[max(0, item_amount-1)])
    print(grid[item_amount])
    _weight = 0
    for value, weight in enumerate(weightss):
        if _weight != weight:
            _weight = weight
            print((value, weight))
```

```python
solution_value = grid[items, capacity]
solution_weight = 0
taken = []
k = capacity
for item in range(items, 0, -1):
    if grid[item][k] != grid[item - 1][k]:
        taken.append(item - 1)
        k -= weights[item - 1]
        solution_weight += weights[item-1]

print(solution_value/100)
print(solution_weight/100)
print(taken)
```
