```python
class RandNumGraph(object):
    
    def __init__(self):
        self.dataset = Queue(5)
        self.threads = []
        
        for x in range(5):
            self.dataset.put(randbelow(100))
            
    def newData(self):
        data = self.dataset.get(block=True, timeout=3)
        self.dataset.put(randbelow(100), block=True, timeout=3)
        return data
```
