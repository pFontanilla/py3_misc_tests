## Fill GroupID Column in Live TCIS DF with Group Part IDs; Groups have blank in TargetPartID but is in own PartID

```python
test_new_tcis_df["TargetPartId"] = np.where((test_new_tcis_df["PartType"].str.contains("GROUP")) & test_new_tcis_df["TargetPartId"].eq(""), test_new_tcis_df["PartId"], test_new_tcis_df["TargetPartId"])
# df_exe = df_exe.sort_values(by=["Thales P/N on BOM", "TargetPartId", "PartType"], ascending=[True, True, False])
display(HTML(test_new_tcis_df.to_html()))
```

## Clean data from new TCIS DF; remove extra sources

```python
def clean(group):
    if (any(group["PartType"].str.contains("GROUP"))):
        group = group[group["TargetPartId"].isin(group["PartId"])]
        group = group.sort_values(by=["Thales P/N on BOM", "TargetPartId", "PartType"], ascending=[True, True, False])
#         display(HTML(group.to_html()))
    else:
        print("Floating Sources")
    return group

def clean2(group):
    if (any(group["PartType"].str.contains("GROUP"))):
        group = group[group["GroupPID"].isin(group["ThalesPID"])]
        group = group.sort_values(by=["ThalesPNBOM", "GroupPID", "PartType"], ascending=[True, True, False])
#         display(HTML(group.to_html()))
    else:
        print("Floating Sources")
    return group

test_new_tcis_df = test_new_tcis_df.groupby("Thales P/N on BOM", as_index=False).apply(func=clean)
test_old_tcis_df = test_old_tcis_df.groupby("ThalesPNBOM", as_index=False).apply(func=clean2)
test_new_tcis_df = test_new_tcis_df.reset_index(drop=True)
```

## Rename columns from Excel to match parquet, QML model column names

## Merge Old data & new data into one 

```python
#join is to combine dfs using index, but in this case we need to use a key
# to specify columns as keys, use merge() function

final_df = pd.merge(test_old_tcis_df, test_new_tcis_df, how="outer", on=["ThalesPNBOM", "ThalesPID", "Live/Local", "GroupPID", "PartType"], suffixes=["Prev", "Updated"], indicator="df")
final_df["dataframe_Source"] = final_df["df"].astype(str).copy()
final_df = final_df.drop(columns=["df"])
final_df = final_df.fillna("")


final_df["Category"] = ""

final_df = final_df[["Category", "ThalesPNBOM", "PartType", "GroupPID", "ThalesPID", "BCNPrev", "BCNUpdated", "DescrPrev",
                     "DescrUpdated", "MPNPrev",
                     "MPNUpdated", "MFRPrev", "MFRUpdated", "AMLPrev", "AMLUpdated",
                     "ApprovalPrev", "ApprovalUpdated", "InternalNotes", "Live/Local", "dataframe_Source"]]

display(HTML(final_df.to_html()))
```

## Sorting correctly, Place Live TCIS row directly above local row

```python
final_df = final_df.sort_values(["ThalesPNBOM", "GroupPID", "PartType", "ThalesPID", "Live/Local"], ascending=[True, True, False, True, True])
display(HTML(final_df.to_html()))
```

## Mandatory Basic Rules

```python
def basic_rules0(group):
    # ThalesPID Level - Modified?
    
   
    # if part is new, all fields blank
    if group.iloc[0, group.columns.get_loc("dataframe_Source")] == "right_only":
        group["Category"] = "New Component"
        
        row = group.iloc[0]
        row["Live/Local"] = "Local"
        group.loc[len(group)] = row
#         display(HTML(group.to_html()))
        return group

    # if part persists in both     
    perfect = True
    
    if group.iloc[0, group.columns.get_loc('BCNPrev')] == group.iloc[0, group.columns.get_loc('BCNUpdated')]:
        group.iloc[1, group.columns.get_loc('BCNUpdated')] = group.iloc[1, group.columns.get_loc('BCNPrev')]
    else:
        perfect = False

    if group.iloc[0, group.columns.get_loc('DescrPrev')] == group.iloc[0, group.columns.get_loc('DescrUpdated')]:
        group.iloc[1, group.columns.get_loc('DescrUpdated')] = group.iloc[1, group.columns.get_loc('DescrPrev')]
    else:
        perfect = False
        
    if group.iloc[0, group.columns.get_loc('MPNPrev')] == group.iloc[0, group.columns.get_loc('MPNUpdated')]:
        group.iloc[1, group.columns.get_loc('MPNUpdated')] = group.iloc[1, group.columns.get_loc('MPNPrev')]
    else:
        perfect = False

    if group.iloc[0, group.columns.get_loc('MFRPrev')] == group.iloc[0, group.columns.get_loc('MFRUpdated')]:
        group.iloc[1, group.columns.get_loc('MFRUpdated')] = group.iloc[1, group.columns.get_loc('MFRPrev')]
    else:
        perfect = False
   
    if group.iloc[0, group.columns.get_loc('AMLPrev')] == group.iloc[0, group.columns.get_loc('AMLUpdated')]:
        group.iloc[1, group.columns.get_loc('AMLUpdated')] = group.iloc[1, group.columns.get_loc('AMLPrev')]
    else:
        perfect = False
   
    if perfect:
        group.iloc[1, group.columns.get_loc('ApprovalUpdated')] = group.iloc[1, group.columns.get_loc('ApprovalPrev')]
        group["Category"] = "No Changes"
    else:
        group["Category"] = "Modified Record"
        
    display(HTML(group.to_html()))
    
    return group

final_df = final_df.groupby(["ThalesPNBOM", "GroupPID", "ThalesPID"], as_index=False).apply(func=basic_rules0)
final_df = final_df.sort_values(["ThalesPNBOM", "GroupPID", "PartType", "ThalesPID", "Live/Local"], ascending=[True, True, False, True, True])
# final_df = final_df.groupby(["ThalesPNBOM", "GroupPID"], as_index=False).apply(func=basic_rules1)

display(HTML(final_df.to_html()))
```
